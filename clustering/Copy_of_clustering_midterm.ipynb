## Handle Missing Values

### Subtask:
Impute missing values in 'CREDIT_LIMIT' and 'MINIMUM_PAYMENTS' using an appropriate strategy (e.g., median imputation) to ensure the dataset is complete for model training.


**Reasoning**:
The subtask requires imputing missing values in 'CREDIT_LIMIT' and 'MINIMUM_PAYMENTS' columns using their respective medians. This code block will calculate the median for each column and then fill the NaN values.



```python
median_credit_limit = df['CREDIT_LIMIT'].median()
median_minimum_payments = df['MINIMUM_PAYMENTS'].median()

df['CREDIT_LIMIT'].fillna(median_credit_limit, inplace=True)
df['MINIMUM_PAYMENTS'].fillna(median_minimum_payments, inplace=True)

print(f"Missing values in 'CREDIT_LIMIT' after imputation: {df['CREDIT_LIMIT'].isnull().sum()}")
print(f"Missing values in 'MINIMUM_PAYMENTS' after imputation: {df['MINIMUM_PAYMENTS'].isnull().sum()}")
```

**Reasoning**:
The previous code produced a FutureWarning regarding the use of `inplace=True` with chained assignment. To address this and follow best practices for Pandas, I will modify the `fillna` calls to directly assign the result back to the DataFrame columns, avoiding the `inplace` argument.



```python
median_credit_limit = df['CREDIT_LIMIT'].median()
median_minimum_payments = df['MINIMUM_PAYMENTS'].median()

df['CREDIT_LIMIT'] = df['CREDIT_LIMIT'].fillna(median_credit_limit)
df['MINIMUM_PAYMENTS'] = df['MINIMUM_PAYMENTS'].fillna(median_minimum_payments)

print(f"Missing values in 'CREDIT_LIMIT' after imputation: {df['CREDIT_LIMIT'].isnull().sum()}")
print(f"Missing values in 'MINIMUM_PAYMENTS' after imputation: {df['MINIMUM_PAYMENTS'].isnull().sum()}")
```

## Scale Numerical Features

## Scale Numerical Features

### Subtask:
Standardize all numerical features to have a mean of 0 and a standard deviation of 1. This is crucial for many machine learning algorithms, especially distance-based ones like K-Means, to prevent features with larger scales from disproportionately influencing the results.

### Reasoning:
Feature scaling is a critical preprocessing step for many machine learning algorithms. Algorithms like K-Means calculate distances between data points, and if features have different scales, features with larger values might dominate the distance calculation, leading to biased results. Standardizing features to a mean of 0 and a standard deviation of 1 (Z-score normalization) ensures that all features contribute equally to the distance metrics. This step will prepare the data for subsequent clustering analysis.

**Reasoning**:
To standardize the numerical features, I will first identify the numerical columns, exclude the non-numerical 'CUST_ID', and then apply `StandardScaler` to transform these columns, storing the results back into the DataFrame.



```python
from sklearn.preprocessing import StandardScaler

# Identify numerical columns, excluding 'CUST_ID'
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
if 'CUST_ID' in numerical_cols:
    numerical_cols.remove('CUST_ID')

# Initialize StandardScaler
scaler = StandardScaler()

# Apply StandardScaler to the numerical columns
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

print("Numerical features scaled successfully.")
display(df.head())
```

## Dimensionality Reduction (PCA)

## Dimensionality Reduction (PCA)

### Subtask:
Apply Principal Component Analysis (PCA) to reduce the dimensionality of the dataset. This can help in visualizing the data and potentially improve the performance of clustering algorithms by removing noise and multicollinearity.

#### Instructions
1. Import the `PCA` class from `sklearn.decomposition`.
2. Initialize a `PCA` object. A common approach is to set `n_components` to a small number (e.g., 2 or 3) for visualization purposes, or a value that explains a significant portion of the variance (e.g., 0.95 for 95% variance).
3. Apply PCA to the scaled numerical features in the DataFrame `df` (excluding the 'CUST_ID' column) using the `.fit_transform()` method.
4. Store the resulting principal components in a new DataFrame, for example, `df_pca`.
5. Print the explained variance ratio for each principal component to understand how much variance each component captures.

### Reasoning:
I will initialize PCA with `n_components=2` to reduce the dataset's dimensionality to two principal components. This choice is ideal for 2D visualization of clusters and often captures a substantial amount of variance, making the data more manageable for clustering algorithms. After applying PCA, I'll create a new DataFrame `df_pca` to hold the components and print the explained variance ratio to assess how much information is retained by these two components.

**Reasoning**:
Now that the instructions and reasoning for PCA have been laid out, I will proceed with implementing the PCA by importing the necessary library, applying PCA to the scaled data, and then creating a new DataFrame for the principal components as instructed.



```python
from sklearn.decomposition import PCA

# Exclude 'CUST_ID' column from PCA if it exists
X = df.drop('CUST_ID', axis=1, errors='ignore')

# Initialize PCA with 2 components for visualization
pca = PCA(n_components=2)

# Apply PCA to the scaled data
X_pca = pca.fit_transform(X)

# Create a new DataFrame with the principal components
df_pca = pd.DataFrame(data = X_pca, columns = ['principal_component_1', 'principal_component_2'])

print("PCA applied successfully. First 5 rows of principal components:")
display(df_pca.head())

print("Explained variance ratio for each principal component:")
print(pca.explained_variance_ratio_)
```

## Determine Optimal Number of Clusters (Elbow Method)

### Subtask:
Use the Elbow Method to identify the optimal number of clusters (K) for the K-Means algorithm. This involves plotting the within-cluster sum of squares (WCSS) against the number of clusters and looking for the 'elbow' point.


**Reasoning**:
To determine the optimal number of clusters using the Elbow Method, I will iterate through a range of cluster numbers, fit a K-Means model for each, calculate the WCSS, and then plot these values to identify the 'elbow' point.



```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Initialize an empty list to store WCSS values
wcss = []

# Define a range of possible cluster numbers
# It's common to start from 1 and go up to a reasonable number, e.g., 10 or 15.
for i in range(1, 11): # Testing k from 1 to 10
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(df_pca)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method graph
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.grid(True)
plt.show()

print("Elbow method plot generated successfully.")
```

## Train K-Means Clustering Model

### Subtask:
Train a K-Means clustering model using the reduced-dimension data and the optimal number of clusters determined by the Elbow Method. Assign each data point to its respective cluster.


### Reasoning:
Based on the Elbow Method plot, the 'elbow' (the point of diminishing returns for adding more clusters) appears to be around K=4. Therefore, I will choose 4 as the optimal number of clusters for the K-Means model. I will then initialize the K-Means model with this number, fit it to the PCA-reduced data, and assign the resulting cluster labels back to the `df_pca` DataFrame for further analysis and visualization.

**Reasoning**:
Now that the optimal number of clusters (K=4) has been identified, I will initialize and train the K-Means model on the PCA-reduced data and then assign the cluster labels to the `df_pca` DataFrame as a new column, as per the instructions.



```python
from sklearn.cluster import KMeans

# Optimal number of clusters based on Elbow Method
optimal_k = 4

# Initialize KMeans model
kmeans = KMeans(n_clusters=optimal_k, init='k-means++', random_state=42, n_init=10)

# Fit KMeans to the PCA-reduced data
kmeans.fit(df_pca)

# Assign cluster labels to df_pca
df_pca['cluster_label'] = kmeans.labels_

print(f"K-Means model trained with {optimal_k} clusters successfully.")
print("First 5 rows of df_pca with cluster labels:")
display(df_pca.head())
```
